```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
```

```python
df = pd.read_csv('Mall_Customers.csv')
df = df.iloc[: , 3:]
df.head()
```

```python
df.shape
```

```python
from sklearn.cluster import KMeans
reg = KMeans(n_clusters=3)
```

```python
X = df.iloc[:,:]
```

```python
reg.fit(X)
```

```python
centroids = reg.cluster_centers_
centroids
```

```python
reg.labels_
```

```python
reg.labels_ == 1,0
```

```python
X.iloc[reg.labels_ == 2,0]
```

```python
centroids[: , 1]
```

```python
plt.figure(figsize=(5,3))
plt.scatter(X.iloc[reg.labels_ == 0, 0], X.iloc[reg.labels_ == 0, 1],c='red')
plt.scatter(X.iloc[reg.labels_ == 1, 0], X.iloc[reg.labels_ == 1, 1],c='green')
plt.scatter(X.iloc[reg.labels_ == 2, 0], X.iloc[reg.labels_ == 2, 1],c='blue')
plt.scatter(centroids[: , 0] , centroids[: , 1],color='black',marker='*')
plt.show()
```

**Best_K[clusters]**

```python
from sklearn.cluster import KMeans
```

```python
k_values = np.arange(2,15)
k_values
```

```python
wcss_values = []
for i in k_values:
  reg = KMeans(n_clusters=i)
  reg.fit(X)
  wcss_values.append(reg.inertia_)

wcss_values
```

```python
plt.figure(figsize=(5,3))
plt.title('Best_k_value')
plt.xlabel('k_value')
plt.ylabel('wcss value for each k_value')
plt.plot(k_values,wcss_values,c='r',marker='*')
plt.show()
```
